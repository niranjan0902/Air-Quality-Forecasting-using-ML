# Data Proccessing

```python
#the data is not in proper manner hence we use sep feature to separate the data by ","
AQD= pd.read_csv("AirQualityUCI.csv",sep=";")
AQD
```

```python
#we have "," insted of decimal point in same features so we have to replace it 
AQD= pd.read_csv("AirQualityUCI.csv",sep=";",decimal=",")
AQD
```

```python
#Removing last 2 Unneccesary features
AQD = AQD.iloc[:,:-2]
AQD
```

```python
#Checking size of the dataset
AQD.shape
```

```python
#there are some missing values in the dataset so we have to remove that rows
AQD = AQD.head(9357)
AQD
```

```python
AQD.info()
```

```python
#checking if there are null values 
AQD.isnull().sum()
```

```python
#this show that there are no missing values in dataset but in actual missing values are tagged with "-200"
#as per USI reposotory documentation.
```

```python
#Counting the number of times -200 appears in data.
AQD.isin([-200]).sum(axis=0)
```

```python
#Handling missing values 
#fisrt we will replace -200 with NaN
AQD = AQD.replace(to_replace = -200,value = np.NaN)
```

```python
AQD.isnull().sum()
#this shows the actual number of missing values
```

```python
#cheaking the mean value of each feature to replace with missing values 
AQD.mean()
```

```python
#Replacing missing values with mean values
AQD = AQD.fillna(AQD.mean())
```

# **Forcasting with FB Prophet algorithm**

FB Prophet Documentation : (https://facebook.github.io/prophet/docs/quick_start.html)

The input to Prophet is always a dataframe with two columns: ds and y. The ds (datestamp) column should be of a format expected by Pandas, ideally YYYY-MM-DD for a date or YYYY-MM-DD HH:MM:SS for a timestamp. The y column must be numeric, and represents the measurement we wish to forecast.

```python
# Converting the date from dd/mm/yyyy to yyyy/mm/dd format

date_info = pd.to_datetime(AQD['Date'])
print(date_info)
```

```python
#we are replacing "." with ":" from time column
time_info = AQD["Time"]
time_info
```

```python
time_info = time_info.apply(lambda x:x.replace(".",":"))
time_info
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#combining 2 series of pandas Dataframe 
date_time = pd.concat([date_info,time_info],axis=1)
date_time
```

```python
date_time.shape
```

```python
#combine both the date and time 
date_time['ds'] = date_time['Date'].astype("str")+' '+date_time["Time"].astype("str")
date_time
```

```python
#we have to convert "ds" from object datatype to datetime format
```

```python
#create empty dataframe
data = pd.DataFrame()
```

```python
data["ds"] = pd.to_datetime(date_time["ds"])
```

```python
data.head()
```

```python
#adding y from main dataset

data["y"] = AQD['RH']
data.head()
```

```python
! pip install prophet
```

```python
from prophet import Prophet
```

```python
#Training the Prophet model
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
